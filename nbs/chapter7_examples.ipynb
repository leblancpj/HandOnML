# Chapter 7: Ensemble Learning and Random Forests 

This chapter is about combining (weak, indpendent) classifiers to get a strong classifier.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
moons = make_moons(n_samples=1000, noise=0.4, random_state=42)
X = moons[0]
y = moons[1]
```

```python
# plot
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X[:,0],X[:,1], c=y)
plt.show();
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=42)
```

## Voting Classifiers 

Note below you can speficy 'hard' or 'soft' for the voting classifier's **voting** parameter.

```python
log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC(probability=True)

voting_clf = VotingClassifier(
    estimators = [('lr',log_clf),('rf',rnd_clf),('svc',svm_clf)],
    voting = 'soft')
voting_clf.fit(X_train, y_train)
```

Let's look at each classifier's accuracy on the test set:

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

## Bagging and Pasting 

"Bagging" stands for "Bootstrap Aggregation".  It means that you resample your raw data set **with** replacement.  "Pasting is the same but it means you resample your raw data set **without** replacement. <br></br>
Just like voting classifiers, the final prediction is typically the statistical <i>mode</i> (most frequent prediction) from all of the classifiers (average for regression problems).

### Bagging and Pasting in Scikit-Learn 

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=500, bootstrap=True, n_jobs=-1
)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

### Out-of-Bag Evaluation

Using boostrap=True, roughly 63% of training instances will be sampled by each predictor.  The other 37% of instances are called the "out-of-bag" (oob) instances.  Note that they are not the same 37% for all predictors.

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    bootstrap=True, n_jobs=-1, oob_score=True)
bag_clf.fit(X_train, y_train)
bag_clf.oob_score_
```

```python
# confirm with accuracy_score
from sklearn.metrics import accuracy_score

y_pred = bag_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

### Random Forests 

Random forests are just optimized versions of what we've done above.  BaggingClassifiers are still useful if you want to bag something other than a DecisionTreeClassifier.

```python
from sklearn.ensemble import RandomForestClassifier

rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1)
rnd_clf.fit(X_train, y_train)

y_pred_rf = rnd_clf.predict(X_test)
```

This is again adding more bias in exchange for reduced variance.  Use Scikit-learn's ExtraTreeClassifier class.  It has the same API as the RandomForestClassifier.

### Feature Importance

Scikit-learn's RandonForestClassifier measures feature importance by averaging how much a given feature reduces impurity across all the trees in the forest.  This is a weighted average, where the weighting is equal to the number of samples that are associated with the node in question.

```python
# feature importance for the Iris dataset
from sklearn.datasets import load_iris

iris = load_iris()
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=-1)
# You don't have train/test split because of oob!
rnd_clf.fit(iris['data'],iris['target'])
# loop through features and give feature importance
for name, score in zip(iris['feature_names'], rnd_clf.feature_importances_):
    print(name, score)
```

## Boosting

<i>Hypothesis Boosting</i> is any method that combines a number of weak learners into a strong learner.  Typically, train predictors sequentially, with each subsequent predictor trying to correct its predecessor.
<br></br>
Most popular methods are <i>AdaBoost</i> and <i>Gradient Boosting</i>.

### AdaBoost 

AdaBoost starts with a base classifier (say, a Decision tree).  Each next predictor puts more emphasis (increases the weight factor) on the training instances that the first classifier underfit (misclassified).  And so on.  In this way, the predictors focus more and more on hard cases.  

The following code trains an AdaBoostClassifier on 200 Decision Stumps (Decision tree with max_depth=1).  This is the default base estimator for AdaBoostClassifier class.

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=200,
    algorithm="SAMME.R", learning_rate=0.5)
ada_clf.fit(X_train, y_train)
```

### Gradient Boosting 

Similar to AdaBoost in that it work sequentially by adding predictors to an ensemble (each one correcting its predecessor). Instead of tweaking the weights for each successive predictor, it tries to fit a new predictor on the <i>residual errors</i> made by the previous predictor.
<br></br>
When doing simple regression with a DecisionTree as a base estimator, this is called <i>Gradient Boosted Regression Trees (GBRT)</i>.

```python
#use a noisy quadratic data set.
m = 500
X = 6. * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m,1)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg1 = DecisionTreeRegressor(max_depth=2)
tree_reg1.fit(X, y)
```

Now, train second tree regressor on the residual errors made by the first.

```python
y2 = y - tree_reg1.predict(X)
tree_reg2 = DecisionTreeRegressor(max_depth=2)
tree_reg2.fit(X, y2)
```

Do it again for a third tree

```python
y3 = y2 - tree_reg2.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2)
tree_reg3.fit(X, y3)
```

Now, we have an ensemble with 3 trees.  Make predictions by adding up the predictions of each tree.

```python
X_new = np.linspace(-3,3,100)
y_pred = sum(tree.predict(X) for tree in (tree_reg1, tree_reg2, tree_reg3))
```

You can easily recreate the above using Scikit Learn's GradientBoostingRegressor class.

```python
from sklearn.ensemble import GradientBoostingRegressor

gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3, learning_rate=1.0)
gbrt.fit(X, y.ravel())
```

How do you determine the optimal number of trees?  If you have too few, your results may underfit and be biased. To many will lead to overfitting.  One way to proceed is to use early <i>early stopping</i>.  To implement we can use the **staged_predict()** method which returns an iterator over the predictions made by an ensemble at each stage of training.

```python
from sklearn.metrics import mean_squared_error
# split data
X_train, X_test, y_train, y_test = train_test_split(X, y)
# create models using too many tree
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=120)
# fit to training split
gbrt.fit(X_train, y_train.ravel())
# get staged predictions 
error = [mean_squared_error(y_test, y_pred) for y_pred in gbrt.staged_predict(X_test)]
# find index (number of trees) where the test error is lowest
bst_n_estimators = np.argmin(error)
print("Best Estimator index = {}".format(bst_n_estimators))
# retrain using optimal number of trees
gbrt_best = GradientBoostingRegressor(max_depth=2, n_estimators = bst_n_estimators)
gbrt_best.fit(X_train, y_train.ravel())
```

An alternative implementation is to actually stop when the error rate starts to go up (instead of training will all of the trees and then looking back for the minimum error point).  Do this by setting warn_start=True in the GBRT.  This causes Scikit Learn to keep existing trees when the fit method is called (I guess by default, it tosses them).

```python
gbrt = GradientBoostingRegressor(max_depth=2, warm_start=True)
min_val_error = float('Inf')
error_going_up = 0
for n_estimators in range(1,120):
    gbrt.n_estimators = n_estimators
    gbrt.fit(X_train, y_train.ravel())
    y_pred = gbrt.predict(X_test)
    error = mean_squared_error(y_test, y_pred)
    if (error <= min_val_error):
        min_val_error = error
        error_going_up = 0
    else:
        error_going_up += 1
        if error_going_up == 5:
            break # early stopping
```
