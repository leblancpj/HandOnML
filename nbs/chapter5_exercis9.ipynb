## Imports

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC, SVR
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix
import warnings
warnings.simplefilter('ignore')
```

## Load Data

```python
boston = load_boston()
```

```python
boston['feature_names']
```

```python
X = boston['data']
y = boston['target']
```

```python
print(X.shape)
print(y.shape)
```

## Train / Test Split 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Train a regressor 

```python
svm_reg = Pipeline([
    ("scaler", StandardScaler()),
    ("reg", SVR(kernel='linear')),
])

svm_reg.fit(X_train, y_train)
```

Let's use mean squared error to evaluate fit on training data.

```python
from sklearn.metrics import mean_squared_error
predict_train = svm_reg.predict(X_train)
train_mse = mean_squared_error(y_train, predict_train)
train_rmse = np.sqrt(train_mse)
train_rmse
```

Do the same thing with the test set.

```python
predict_test = svm_reg.predict(X_test)
test_mse = mean_squared_error(y_test, predict_test)
test_rmse = np.sqrt(test_mse)
test_rmse
```

## Use Cross Validation for better evaluation

```python
scores = cross_val_score(svm_reg, X_test, y_test, cv=10, scoring='neg_mean_squared_error')
svm_reg_scores = np.sqrt(-scores)
```

## Fine-tune using GridSearchCV

```python
svm_reg.get_params().keys()
```

```python
from sklearn.model_selection import GridSearchCV

params = [
    {"reg__C":[0.01,0.1,1,10,100], "reg__kernel":["linear"]},
    {"reg__C":[0.01,0.1,1,10,100], "reg__kernel":["rbf"]},
    {"reg__C":[0.01,0.1,1,10,100], "reg__kernel":["poly"]},
]
```

```python
grid_search = GridSearchCV(svm_reg, param_grid=params, cv=5, scoring="neg_mean_squared_error")
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
cv_results = grid_search.cv_results_
for mean_score, params in zip(cv_results["mean_test_score"], cv_results["params"]):
    print(np.sqrt(-mean_score), params)
```
