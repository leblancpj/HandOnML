Train and fine-tune a Decision Tree for the moons dataset. <b></b>

    <li>Use grid search with cross-validation (with the help of the GridSearchCV class) to find good hyperparameter values for a DecisionTreeClassifier.  Hint: try various values for max_leaf_nodes.</li>
    <li>Train it on the full training set using these hyperparametersm and measure your model's performance on the test set.  You should get roughly 85% to 87% accuracy.</li>

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

## Load Dataset 

```python
moons = make_moons(n_samples=10000, noise=0.4, random_state=42)
X = moons[0]
y = moons[1]
print(X[:5])
print(y[:5])
```

```python
# plot
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X[:,0],X[:,1], c=y)
plt.show();
```

## Create Training and Test Set 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=42)
```

## Create a Classifier

```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
```

## Use Grid Search 

```python
# define a range of parameters to search
params = [
    {'max_leaf_nodes':range(2,10,1)}
]
grid_search = GridSearchCV(tree_clf,param_grid=params,cv=10,scoring="neg_mean_squared_error")
grid_search.fit(X_train, y_train)
```

```python
cv_results = grid_search.cv_results_
for mean_score, params in zip(cv_results["mean_test_score"], cv_results["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
grid_search.best_estimator_
```

It appears that thers is no improvement using max_leaf_nodes greater than 4.  So that's what we will use.

## Re-train on full training set using best model 

```python
tree_clf_best = DecisionTreeClassifier(max_leaf_nodes=4, random_state=42)
tree_clf_best.fit(X_train, y_train)
```

## Make predictions

```python
y_pred = tree_clf_best.predict(X_train)
```

## Check accuracy of predictions

```python
# precision / recall
from sklearn.metrics import precision_score, recall_score, f1_score
print(precision_score(y_pred, y_train, average='weighted'))
print(recall_score(y_pred, y_train, average='weighted'))
print(f1_score(y_pred, y_train, average='weighted'))
```

## Repeat using Test Set 

```python
y_pred_2 = tree_clf_best.predict(X_test)
print(f1_score(y_pred_2, y_test, average='weighted'))
```

    <li>Now comes the magic.  For each test set instance, generate the predictions of the 1,000 Decision Trees and keep only the most frequent prediction (you can use SciPy's mode() function for this). This gives you <i>majority-vote predictions</i> over the test set.</li>
    <li>Evaluate these predictions on the test set: you should obtain a <i>slightly higher</i> accuracy than your first model (about 0.5 to 1.5% higher).  Congratulations, you have trained a Random Forest Classifier!</li>

## Generate 1000 Training Subsets

```python
from sklearn.model_selection import ShuffleSplit
ss = ShuffleSplit(n_splits=1000, train_size=100, random_state=42)
```

## Train a tree on each subset 

```python
from sklearn.metrics import accuracy_score
```

```python
# create a list to hold the trees
tree_clf_list = []
f1_scores = []
acc_scores = []
for train_index, test_index in ss.split(X_train):
    X_train_subset, y_train_subset = X_train[train_index], y_train[train_index]
    X_test_subset, y_test_subset = X_train[test_index], y_train[test_index]
    tree_clf_subset = DecisionTreeClassifier(max_leaf_nodes=4, random_state=42)
    tree_clf_subset.fit(X_train_subset, y_train_subset)
    tree_clf_list.append(tree_clf_subset)
    y_pred_subset = tree_clf_subset.predict(X_test_subset)
    acc_scores.append(accuracy_score(y_test_subset, y_pred_subset))
    f1_scores.append(f1_score(y_test_subset, y_pred_subset, average='weighted'))
    
print(y_pred_subset[:10])
print(acc_scores[:10])
print(f1_scores[:10])
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from collections import Counter
clf_preds = np.ndarray((len(X_test), len(tree_clf_list)))
majority_pred = np.ndarray(len(X_test))
print(clf_preds.shape)
votes = []
# Loop through classifiers
clf_count = 0
for clf in tree_clf_list:
    #print (clf)
    pred = clf.predict(X_test)
    #print(clf_count, pred)
    clf_preds[:, clf_count] = pred
    #votes.append(pred.item())
    clf_count += 1
    #if (clf_count > 2):
    #    break
        
#print(clf_preds[0,:])

# Loop through predictions
instance_counter = 0
for sample in clf_preds:
    #print(sample[:5])
    #break
    c = Counter(sample)
    majority = c.most_common(2)
    majority_pred[instance_counter] = majority[0][0]
    #print ("majority / count: ", majority)
    instance_counter += 1
    
# score majority prediction
#print(y_test[:10])
#print(majority_pred[:])
majority_vote_score = accuracy_score(y_test, majority_pred)
print(majority_vote_score)
```
